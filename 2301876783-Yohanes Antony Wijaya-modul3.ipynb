OBJECT AND CLASS

```python
class ident:
  def __init__(self, name, age):
    self.name = name
    self.age = age

p1 = ident("Yohan", 20)

print(p1.name)
print(p1.age)
```

```python
class ident:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def help(self):
    print("Hello my name is " + self.name)

p1 = ident("Yohan", 20)
p1.help()
```
